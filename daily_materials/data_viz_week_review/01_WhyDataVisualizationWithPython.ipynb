<h2>Why use python for data visualization</h2>
If you have ever graphed anything using Excel or a business intelligence tool like Power BI or Tableau, you might be wondering what is the point of using a programming language like Python. There are a couple reasons.

<b>Automation</b>

One task people often use Python for is automation. If you are a spreadsheet user, you may have been hours renaming files or updating spreadsheet cells before you were able to visualize your data. With Python, you can automate renaming of files and even the creation of graphs in Excel using Python. If you want to learn more about automating things in Python, you can read the free book <a href="https://automatetheboringstuff.com/">Automate the Boring Stuff with Python</a>.
