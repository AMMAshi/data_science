<h2>Import Libraries</h2>

```python
import pandas as pd

# `plt` is an alias for the `matplotlib.pyplot` module
import matplotlib.pyplot as plt
```

<h2>plot method</h2>
The plot method accepts various python data structures including lists, tuples, and NumPy arrays. While matplotlib is always improving, at the moment you cannot plot pandas Dataframes using plt.plot(you can plot directly from pandas DataFrames though). As a reminder, if you dont know what a method accepts, you can use the in built-in function help

```python
# Not the prettiest plot
plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
```

```python
# You can also plot another line on the same graph
plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
plt.plot([10, 20, 30, 40, 50], [100, 400, 900, 1600, 2500])
```

<h2>Choose figure style</h2>

The default style is not the most aesthetically pleasing.

You can use `plt.style.available` to see different aesthetic styles available for your figures. 


```python
plt.style.available
```

```python
# We are currently using default plot 
# so this will not change much about our plot
plt.style.use('classic')
plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
```

```python
plt.style.use('fivethirtyeight')

plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
```

```python
plt.style.use('ggplot')

plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
```

```python
plt.style.use('tableau-colorblind10')

plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
```

```python
plt.style.use('seaborn')

plt.plot([10, 20, 30, 40, 50], [5, 10, 15, 20, 25])
```
