<h2>Import libraries</h2>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

<h2>Load the dataset</h2>
This is a subset of the <a href="https://www.kaggle.com/heesoo37/120-years-of-olympic-history-athletes-and-results">120 years of Olympic history dataset</a> on Kaggle. 

```python
# Keep in mind that the filename will likely be different on your computer (especially later in course)
filename = 'data/athleteEventsNoPersonal.csv'
df = pd.read_csv(filename)
df.head()
```

Getting data into a form suitable for graphing. This is covered in previous sections. 

```python
# get number of olympians per year
numberUniqueYear = df.groupby(['Year'])['ID'].nunique()
```

```python
# this is a pandas Series
numberUniqueYear
```

```python
# Can convert to NumPy array or list

uniqueYears = np.array(numberUniqueYear.index)
uniqueYears
```

```python
numOlympian = np.array(numberUniqueYear.values)
numOlympian
```

<h2>Set title, labels, and limits</h2>
This section covers how to add limits, labels, and titles to your graph. I highly encourage you to play with adding and removing these on your own as it is a great way to learn. Note that all of these skills will come in handy when you are trying to validate machine learning algorithms. 

```python
# Make the style by default seaborn
plt.style.use('seaborn')
```

```python
# Plot graph showing how it would look without defaults
plt.plot(uniqueYears, numOlympian, c= 'k')
```

<b>MATLAB-style</b>

```python
# Set xlim and ylim
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.xlim(left=1892,right=2020)
plt.ylim(bottom=0,top=1800)
```

```python
# Set xlabel and ylabel
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.xlim(left=1892,right=2020)
plt.ylim(bottom=0,top=1750)
plt.xlabel('Year')
plt.ylabel('Number of Olympians')
```

```python
# Set title
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.xlim(left=1892,right=2020)
plt.ylim(bottom=0,top=1750)
plt.xlabel('Year')
plt.ylabel('Number of Olympians')
plt.title('Number of Olympians Over the Years')
```

```python
# Change fontsize
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.xlim(left=1892,right=2020)
plt.ylim(bottom=0,top=1750)
plt.xlabel('Year', fontsize = 16)
plt.ylabel('Number of Olympians', fontsize = 16)
plt.title('Number of Olympians Over the Years', fontsize = 16)
```

<b>Object-oriented</b>

```python
# Set xlim and ylim
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(uniqueYears, numOlympian, c= 'k');
axes.set_xlim(left=1892,right=2020)
axes.set_ylim(bottom=0,top=1800)
```

```python
# Set xlabel and ylabel
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(uniqueYears, numOlympian, c= 'k');
axes.set_xlim(left=1892,right=2020)
axes.set_ylim(bottom=0,top=1800)
axes.set_xlabel('Year')
axes.set_ylabel('Number of Olympians')
```

```python
# Set title
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(uniqueYears, numOlympian, c= 'k');
axes.set_xlim(left=1892,right=2020)
axes.set_ylim(bottom=0,top=1800)
axes.set_xlabel('Year')
axes.set_ylabel('Number of Olympians')
axes.set_title('Number of Olympians Over the Years')
```

```python
# Change fontsize
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(uniqueYears, numOlympian, c= 'k');
axes.set_xlim(left=1892,right=2020)
axes.set_ylim(bottom=0,top=1800)
axes.set_xlabel('Year', fontsize = 16)
axes.set_ylabel('Number of Olympians', fontsize = 16)
axes.set_title('Number of Olympians Over the Years', fontsize = 16)
```
