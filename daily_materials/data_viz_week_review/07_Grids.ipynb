<h2>Import libraries</h2>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

<h2>Load the dataset</h2>
This is a subset of the <a href="https://www.kaggle.com/heesoo37/120-years-of-olympic-history-athletes-and-results">120 years of Olympic history dataset</a> on Kaggle. 

```python
# Keep in mind that the filename will likely be different on your computer (especially later in course)
filename = 'data/athleteEventsNoPersonal.csv'
df = pd.read_csv(filename)
df.head()
```

Getting data into a form suitable for graphing. This is covered in previous sections. 

```python
# get number of olympians per year
numberUniqueYear = df.groupby(['Year'])['ID'].nunique()
```

```python
# this is a pandas Series
numberUniqueYear
```

```python
# Can convert to NumPy array or list

uniqueYears = np.array(numberUniqueYear.index)
uniqueYears
```

```python
numOlympian = np.array(numberUniqueYear.values)
numOlympian
```

<h2>Grids</h2>
This section covers how to add grids to your plots. While you can use different default styles (plt.style.use) to have your plots automatically have grids, it is good to know how to edit everything in your plots. 

```python
# Plot graph showing how it would look with defaults
plt.plot(uniqueYears, numOlympian, c= 'k')
```

<b>MATLAB-style</b>

```python
# Add a grid
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.grid()
```

```python
# only horizontal grid lines
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.grid(axis = 'y')
```

```python
# only vertical grid lines
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.grid(axis = 'x')
```

```python
# change color of grid lines, transparency, and linestyle
plt.plot(uniqueYears, numOlympian, c= 'k')
plt.grid(c = 'g', 
         alpha = .9,
         linestyle = '-')
```

<b>Object-oriented</b>

```python
# Add a grid
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(uniqueYears, numOlympian, c= 'k');
axes.grid()
```

```python
# only horizontal grid lines
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(uniqueYears, numOlympian, c= 'k');
axes.grid(axis = 'y')
```

```python
# only vertical grid lines
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(uniqueYears, numOlympian, c= 'k');
axes.grid(axis = 'x')
```
